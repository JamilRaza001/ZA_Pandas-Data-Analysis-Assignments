```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('country_data.csv')
```

```python
data.head()
```

```python
life_expectancy = data['Life Expectancy']
life_expectancy
```

```python
life_expectancy_quartiles  = np.quantile(life_expectancy, [0.25, 0.5, 0.75, 1])
print(life_expectancy_quartiles)
```

```python

# Histogram plot ko variable 'ax' mein store karte hain:
ax = sns.histplot(life_expectancy)  # yahan bins ko customize kar sakte hain, misal ke taur par 10 bins

# Har bar (bin) ke liye annotation add karna:
for patch in ax.patches:
    # patch.get_height() se har bar ka height milta hai, jo ke bin count hota hai.
    # patch.get_x() se bar ka starting position aur patch.get_width() se bar ki width milti hai.
    # Yeh calculate karta hai ke annotation kahaan center mein aa jaye.
    ax.annotate(format(patch.get_height(), '.0f'),
                (patch.get_x() + patch.get_width() / 2, patch.get_height()),
                ha='center', va='center', 
                xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
gdp = data['GDP']
```

```python
median_gdp = np.quantile(a=gdp,q=0.5)
print(median_gdp)
```

```python
low_gdp = data[data['GDP'] <= median_gdp]
low_gdp.info()
```

```python
high_gdp = data[data['GDP'] > median_gdp]
high_gdp
```

```python
low_gdp_quartiles = np.quantile(low_gdp['Life Expectancy'],[0.25,0.50,0.75,1])
low_gdp_quartiles
```

```python
high_gdp_quartiles = np.quantile(high_gdp['Life Expectancy'],[0.25,0.50,0.75,1])
high_gdp_quartiles
```

```python
sns.histplot(high_gdp["Life Expectancy"], alpha = 0.5, label = "High GDP")
sns.histplot(low_gdp["Life Expectancy"], alpha = 0.5, label = "Low GDP")
plt.legend()
plt.show()
```
